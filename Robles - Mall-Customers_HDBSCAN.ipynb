```python
# Import Required Libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import hdbscan

%matplotlib inline
sns.set_context('poster')
sns.set_style('white')
sns.set_color_codes()
plot_kwds = {'alpha': 0.5, 's': 80, 'linewidths': 0}
```

```python
test_data = pd.read_csv('Mall_Customers.csv', encoding='Latin-1') 
test_data.head()
```

```python
test_data = test_data.iloc[:,[3,4]].values
```

```python
test_data
```

```python
plt.scatter(test_data[:,0], test_data[:,1], color='b', **plot_kwds)
```

```python
clusterer_size3 = hdbscan.HDBSCAN(min_cluster_size=3, gen_min_span_tree=True)
clusterer_size3.fit(test_data)
```

```python
clusterer_size5 = hdbscan.HDBSCAN(min_cluster_size=5, gen_min_span_tree=True)
clusterer_size5.fit(test_data)
```

```python
clusterer_size9 = hdbscan.HDBSCAN(min_cluster_size=9, gen_min_span_tree=True)
clusterer_size9.fit(test_data)
```

```python
clusterer_size3.minimum_spanning_tree_.plot(edge_cmap='viridis',
                                      edge_alpha=0.6,
                                      node_size=80,
                                      edge_linewidth=2)
```

```python
clusterer_size5.minimum_spanning_tree_.plot(edge_cmap='viridis',
                                      edge_alpha=0.6,
                                      node_size=80,
                                      edge_linewidth=2)
```

```python
clusterer_size9.minimum_spanning_tree_.plot(edge_cmap='viridis',
                                      edge_alpha=0.6,
                                      node_size=80,
                                      edge_linewidth=2)
```

```python
clusterer_size3.condensed_tree_.plot()
```

```python
clusterer_size5.condensed_tree_.plot()
```

```python
clusterer_size9.condensed_tree_.plot()
```

```python
clusterer_size3.condensed_tree_.plot(select_clusters=True, selection_palette=sns.color_palette())
```

```python
clusterer_size5.condensed_tree_.plot(select_clusters=True, selection_palette=sns.color_palette())
```

```python
clusterer_size9.condensed_tree_.plot(select_clusters=True, selection_palette=sns.color_palette())
```

```python
palette = sns.color_palette()
cluster_colors = [sns.desaturate(palette[col], sat)
                  if col >= 0 else (0.5, 0.5, 0.5) for col, sat in
                  zip(clusterer_size3.labels_, clusterer_size3.probabilities_)]
plt.scatter(test_data.T[0], test_data.T[1], c=cluster_colors, **plot_kwds)
plt.title("HDBSCAN (Cluster Size: 3)")
plt.show()
```
